```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=150):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ... (your existing code to load train and test data) ...

# Get 'Close' column from train and test DataFrames
train_close = data[['Close']].values
test_close = testData[['Close']].values

# Scale the 'Close' values for training and testing data
scaler = MinMaxScaler(feature_range=(0, 1))
train_close_scaled = scaler.fit_transform(train_close)
test_close_scaled = scaler.transform(test_close)  # Scale test data using the same scaler

# ... (rest of your code using train_close_scaled and test_close_scaled) ...

def create_sequences(data, timestep=60):
    X, y = [], []
    for i in range(timestep, len(data)):
        X.append(data[i - timestep:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

timestep = 60
X_train, y_train = create_sequences(train_close_scaled, timestep)
X_test, y_test = create_sequences(test_close_scaled, timestep)  # Use test_close_scaled here
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
#history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=0)

y_pred = model.predict(X_test)
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='red', label='Actual Stock Price')
plt.plot(y_pred_rescaled, color='blue', label='Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")

tolerance = 0.05
percentage_diff = np.abs(y_test_rescaled - y_pred_rescaled) / y_test_rescaled
accuracy = np.mean(percentage_diff <= tolerance) * 100
precision = np.sum(percentage_diff <= tolerance) / len(y_test_rescaled) * 100

print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ========== Load your data ==========
# Assume data and testData are already loaded:
# data = pd.read_csv("train_dataset.csv")
# testData = pd.read_csv("test_dataset.csv")

# Use only the 'Close' column
train_close = data[['Close']].values
test_close = testData[['Close']].values

# ========== Scaling ==========
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_close)
test_scaled = scaler.transform(test_close)

# ========== Sequence Preparation ==========
def prepare_data(data, timestep):
    X, y = [], []
    for i in range(timestep, len(data)):
        X.append(data[i - timestep:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

timestep = 60
X_train, y_train = prepare_data(train_scaled, timestep)
X_test, y_test = prepare_data(test_scaled, timestep)

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# ========== Lasso Regression ==========
lasso = Lasso(alpha=0.001)
lasso.fit(X_train_flat, y_train)
y_pred = lasso.predict(X_test_flat)

# Inverse scaling for evaluation
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_rescaled = scaler.inverse_transform(y_pred.reshape(-1, 1))

# --- Plot Lasso Predictions ---
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='red', label='Actual Stock Price')
plt.plot(y_pred_rescaled, color='blue', label='Predicted Stock Price (Lasso)')
plt.title('Stock Price Prediction using Lasso Regression')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Lasso Metrics ---
tolerance = 0.05
percentage_diff = np.abs(y_test_rescaled - y_pred_rescaled) / y_test_rescaled
accuracy = np.mean(percentage_diff <= tolerance) * 100
precision = np.sum(percentage_diff <= tolerance) / len(y_test_rescaled) * 100
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

print("=== Lasso Regression Metrics ===")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")
print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}%")

# ========== PCA + Lasso ==========
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_flat)
X_test_pca = pca.transform(X_test_flat)

lasso.fit(X_train_pca, y_train)
y_pred_pca = lasso.predict(X_test_pca)
y_pred_pca_rescaled = scaler.inverse_transform(y_pred_pca.reshape(-1, 1))

# --- Plot PCA + Lasso ---
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='red', label='Actual Stock Price')
plt.plot(y_pred_pca_rescaled, color='green', label='Predicted Price (PCA + Lasso)')
plt.title('Stock Price Prediction with PCA + Lasso')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- PCA + Lasso Metrics ---
percentage_diff_pca = np.abs(y_test_rescaled - y_pred_pca_rescaled) / y_test_rescaled
accuracy_pca = np.mean(percentage_diff_pca <= tolerance) * 100
precision_pca = np.sum(percentage_diff_pca <= tolerance) / len(y_test_rescaled) * 100

print("=== PCA + Lasso Metrics ===")
print(f"PCA + Lasso Accuracy: {accuracy_pca:.2f}%")
print(f"PCA + Lasso Precision: {precision_pca:.2f}%")

# ========== PCA-Only Reconstruction ==========
X_test_reconstructed = pca.inverse_transform(X_test_pca)
reconstructed_close_scaled = X_test_reconstructed[:, -1].reshape(-1, 1)
reconstructed_close = scaler.inverse_transform(reconstructed_close_scaled)

# --- Plot PCA Only ---
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, label='Actual Stock Price', color='red')
plt.plot(reconstructed_close, label='PCA Reconstructed Close Price', color='orange')
plt.title('Stock Price Reconstruction using PCA Only')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- PCA-Only Metrics ---
percentage_diff_pca_only = np.abs(y_test_rescaled - reconstructed_close) / y_test_rescaled
accuracy_pca_only = np.mean(percentage_diff_pca_only <= tolerance) * 100
precision_pca_only = np.sum(percentage_diff_pca_only <= tolerance) / len(y_test_rescaled) * 100
mse_pca_only = mean_squared_error(y_test_rescaled, reconstructed_close)
mae_pca_only = mean_absolute_error(y_test_rescaled, reconstructed_close)
r2_pca_only = r2_score(y_test_rescaled, reconstructed_close)

print("=== PCA Only Metrics ===")
print(f"Mean Squared Error: {mse_pca_only:.2f}")
print(f"Mean Absolute Error: {mae_pca_only:.2f}")
print(f"R² Score: {r2_pca_only:.2f}")
print(f"PCA Only Accuracy: {accuracy_pca_only:.2f}%")
print(f"PCA Only Precision: {precision_pca_only:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.utils import check_array

# Simulate dummy stock data for demo (replace this with your actual data loading)
np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=500, freq='D')
prices = np.cumsum(np.random.randn(500)) + 100
data = pd.DataFrame({'Date': dates[:400], 'Close': prices[:400]})
testData = pd.DataFrame({'Date': dates[400:], 'Close': prices[400:]})

# Use only the 'Close' column
train_close = data[['Close']].values
test_close = testData[['Close']].values

# Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_close)
test_scaled = scaler.transform(test_close)

# Sequence Preparation
def prepare_data(data, timestep):
    X, y = [], []
    for i in range(timestep, len(data)):
        X.append(data[i - timestep:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

timestep = 60
X_train, y_train = prepare_data(train_scaled, timestep)
X_test, y_test = prepare_data(test_scaled, timestep)

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Lasso Regression
lasso = Lasso(alpha=0.001)
lasso.fit(X_train_flat, y_train)
y_pred = lasso.predict(X_test_flat)

# Inverse scaling for evaluation
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
y_pred_rescaled = scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()

# Metrics and plot for Lasso
tolerance = 0.05
percentage_diff = np.abs(y_test_rescaled - y_pred_rescaled) / np.maximum(y_test_rescaled, 1e-8)
accuracy = np.mean(percentage_diff <= tolerance) * 100
precision = np.sum(percentage_diff <= tolerance) / len(y_test_rescaled) * 100
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

# PCA + Lasso
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_flat)
X_test_pca = pca.transform(X_test_flat)

lasso.fit(X_train_pca, y_train)
y_pred_pca = lasso.predict(X_test_pca)
y_pred_pca_rescaled = scaler.inverse_transform(y_pred_pca.reshape(-1, 1)).flatten()

percentage_diff_pca = np.abs(y_test_rescaled - y_pred_pca_rescaled) / np.maximum(y_test_rescaled, 1e-8)
accuracy_pca = np.mean(percentage_diff_pca <= tolerance) * 100
precision_pca = np.sum(percentage_diff_pca <= tolerance) / len(y_test_rescaled) * 100

# PCA-Only
X_test_reconstructed = pca.inverse_transform(X_test_pca)
reconstructed_close_scaled = X_test_reconstructed[:, -1].reshape(-1, 1)
reconstructed_close = scaler.inverse_transform(reconstructed_close_scaled).flatten()

percentage_diff_pca_only = np.abs(y_test_rescaled - reconstructed_close) / np.maximum(y_test_rescaled, 1e-8)
accuracy_pca_only = np.mean(percentage_diff_pca_only <= tolerance) * 100
precision_pca_only = np.sum(percentage_diff_pca_only <= tolerance) / len(y_test_rescaled) * 100
mse_pca_only = mean_squared_error(y_test_rescaled, reconstructed_close)
mae_pca_only = mean_absolute_error(y_test_rescaled, reconstructed_close)
r2_pca_only = r2_score(y_test_rescaled, reconstructed_close)

# Store metrics in a DataFrame
metrics = {
    "Lasso": {"MSE": mse, "MAE": mae, "R2": r2, "Accuracy": accuracy, "Precision": precision},
    "PCA + Lasso": {"MSE": "-", "MAE": "-", "R2": "-", "Accuracy": accuracy_pca, "Precision": precision_pca},
    "PCA Only": {"MSE": mse_pca_only, "MAE": mae_pca_only, "R2": r2_pca_only,
                 "Accuracy": accuracy_pca_only, "Precision": precision_pca_only}
}

results_df = pd.DataFrame(metrics).T
print("\n=== Model Evaluation Metrics ===")
print(results_df)
```
